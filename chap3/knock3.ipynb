# 顧客データの成形

```python
customer_join = pd.merge(customer_master, class_master, on="class", how="left")
customer_join = pd.merge(customer_join, campaign_master, on="campaign_id", how="left")
customer_join = customer_join[["customer_id", "name", "gender", "start_date", "end_date", "class", "class_name", "price", "campaign_id", "campaign_name", "is_deleted"]]
len(customer_join)
customer_join
```

```python
customer_join.isnull().sum()
```

# 基礎集計

```python
customer_join.groupby('class_name').count()['customer_id']
```

```python
customer_join.groupby('campaign_name').count()['customer_id']
```

```python
customer_join.groupby('gender').count()['customer_id']
```

```python
customer_join.groupby('is_deleted').count()['customer_id']
```

```python
customer_join["start_date"] = pd.to_datetime(customer_join["start_date"])
customer_start = customer_join.loc[customer_join["start_date"]>pd.to_datetime("20180401")]
len(customer_start)
```

```python
customer_start["start_month"] = customer_start["start_date"].dt.strftime("%Y%m")
customer_start.pivot_table(index="start_month", columns="class_name", aggfunc="size", fill_value=0)
```
